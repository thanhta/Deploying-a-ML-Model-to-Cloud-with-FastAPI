```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Open US Census Income Data which needs to clean up

```python
data_path = "../data/census.csv"
data_df = pd.read_csv(data_path, skipinitialspace=True)
data_df
```

```python
# Find out number of rows and columns of the US Census Income data
data_df.shape
```

```python
# Find out the data types of each columns
data_df.info()
```

```python
# Find out the stastics of each column's data such as: what is the min and max values, mean value, standartd deviation, etc
data_df.describe()
```

### Replace any "?" by None and drop any row with value equals to NA

```python
clean_df = data_df.replace("?", None).dropna()
clean_df
```

### Remove leading amd traling whitespace

```python
clean_df.columns = [x.strip() for x in clean_df.columns]
```

```python
print('sum of null values for each column:\n')
clean_df.isnull().sum()
```

### There is no null value for each column

### Remove duplicate rows if exists

```python
clean_df.drop_duplicates(inplace=True, ignore_index=False)
```

```python
clean_df['salary'].value_counts()
```

### There are 22633 rows which having salary <= 50k and 7506 rows having salary > 50k

### Save clean_df to a new file with name: census_cleaned.csv

```python
clean_data_path = "../data/census_cleaned.csv"
clean_df.to_csv(clean_data_path, index=False)
```
